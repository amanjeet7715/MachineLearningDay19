```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

**x, y, hue : (names of variables in data or vector data, optional)
Inputs for plotting long-form data. See examples for interpretation.**

```python
sns.barplot(x=iris.sepal_length, y=iris.petal_length)
```

```python
sns.barplot(x=iris.sepal_length, y=iris.petal_length, hue=iris.species)
```

**data : (DataFrame, array, or list of arrays, optional)
Dataset for plotting. If x and y are absent, this is interpreted as wide-form. Otherwise it is expected to be long-form.**

```python
sns.barplot(x='sepal_length', y='petal_length', hue='species', data=iris)
```

**orient : (“v” | “h”, optional)
Orientation of the plot (vertical or horizontal). This is usually inferred based on the type of the input variables, but it can be used to resolve ambiguitiy when both x and y are numeric or when plotting wide-form data.**

```python
sns.barplot(x=iris.sepal_length, y=iris.petal_length, hue=iris.species, orient='h')
```

**color : (matplotlib color, optional)
Color for all of the elements, or seed for a gradient palette.**

```python
sns.barplot(x=iris.sepal_length, y=iris.petal_length, hue=iris.species, color='brown')
```

**capsize : (float, optional)
Width of the “caps” on error bars.**

```python
sns.barplot(x=iris.sepal_length, y=iris.petal_length, hue=iris.species, capsize=.2)
```

**dodge : (bool, optional)
When hue nesting is used, whether elements should be shifted along the categorical axis.**

```python
sns.barplot(x=iris.sepal_length, y=iris.petal_length, hue=iris.species, dodge=False)
```

**palette : (palette name, list, or dict)
Colors to use for the different levels of the hue variable. Should be something that can be interpreted by color_palette(), or a dictionary mapping hue levels to matplotlib colors.**

```python
sns.barplot(x='sepal_length', y='petal_length', hue='species', data=iris, palette="Blues_d")
```
